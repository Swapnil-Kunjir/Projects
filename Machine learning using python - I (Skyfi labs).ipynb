```python
#Normalisation of BH data
X=df["LSTAT"].values
Y=df['target'].values
```

```python
x_scaler=MinMaxScaler()
X=x_scaler.fit_transform(X.reshape(-1,1))
X=X[:,-1]
```

```python
y_scaler=MinMaxScaler()
Y=y_scaler.fit_transform(Y.reshape(-1,1))
Y=Y[:,-1]
```

```python
#Error Function for calculating Cost Function i.e. MSE
def error(m,x,c,t):
    N=x.size
    e=sum(((m*x+c)-t)**2)
    return e*1/(2*N)
```

```python
#Splitting Data into Training And Test
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
```

```python
#Update Function to update wieghts of m & c
def update(m,x,c,t,learning_rate):
    grad_m=sum(2*((m*x+c)-t)*x)
    grad_c=sum(2*((m*x+c)-t))
    m=m-grad_m*learning_rate
    c=c-grad_c*learning_rate
    return m,c
```

```python
#Gradient Decent Function For getting best fit Values For m & c using Update and Error Function
def gradient_descent(init_m,init_c,x,t,learning_rate,iterations,error_threshold):
    m=init_m
    c=init_c
    error_values=list()
    mc_values=list()
    for i in range(iterations):
        e=error(m,x,c,t)
        if e<error_threshold:
            print('Error less than the threshold. Stopping gradient descent.')
            break
        error_values.append(e)
        m,c=update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m,c,error_values,mc_values
```

```python

%%time
init_m=0.9
init_c=0
learning_rate=0.001
iterations=250
error_threshold=0.001


m,c,error_values,mc_values=gradient_descent(init_m,init_c,xtrain,ytrain,learning_rate,iterations,error_threshold)
```

```python
#As number of iterations increases, changes in the line are lessnoticable.
#In order to reduce iteration time for the animation it is advisable to choose smaller time
mc_values_anim=mc_values[0:250:5]
```

```python
import matplotlib.animation as manimation; manimation.writers.list()
```

```python
#Visualization of Training Model Fitting
%matplotlib inline

fig,ax=plt.subplots()
ln,=plt.plot([],[],'ro-',animated=True)


def init():
    plt.scatter(xtrain,ytrain,color='g')
    ax.set_xlim(0,1.0)
    ax.set_ylim(0,1.0)
    return ln,

def update_frame(frame):
    m,c=mc_values_anim[frame]
    x1,y1=-0.5,m*(-0.5)+c
    x2,y2=1.5,m*(1.5)+c
    ln.set_data([x1,x2],[y1,y2])
    return ln,

anim=mpl.animation.FuncAnimation(fig,update_frame,frames=range(len(mc_values_anim)),init_func=init,blit=True)

HTML(anim.to_html5_video())
```

```python
#Visualisation of Predicted values
plt.scatter(xtrain,ytrain,color='b')
plt.plot(xtrain,(m*xtrain+c),color='r')
```

```python
#Visualisation of error
plt.plot(np.arange(len(error_values)),error_values)
plt.ylabel('Error')
plt.xlabel('Iterations')
```

```python
#calculate the predictions on the test set as a vectorized operations
predicted=m*xtest+c
```

```python
#compute the MSE for the predicted values on the test data set
mean_squared_error(ytest,predicted)
```

```python
# Put xtest,ytest and predicted values into a single DataFrame so that we can see the predicted values alongside with testing set
p=pd.DataFrame(list(zip(xtest,ytest,predicted)),columns=['x','target_y','predicted_y'])
p.head()
```
